```python
tmp = pd.read_csv('/home/phuongdx/python/GAPO/data/merged_data.csv')
```

```python
tmp.columns = ['post_id', 'gapo_hashtag']
tmp.head(10)
```

```python
content_list = []
image_id_list = []
image_src_list = []
```

```python
for index, row in tmp.iterrows():
    content, image_id, image_src = get_image_id_and_text(row[0])
    content_list.append(content)
    image_id_list.append(image_id)
    image_src_list.append(image_src)
tmp['image_id'] = image_id_list
tmp['content'] = content_list
tmp['image_src'] = image_src_list
tmp.head(10)
```
